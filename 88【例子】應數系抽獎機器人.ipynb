```python
img = Image('images/tgchen_draw.png', width=100)
img2 = Image('images/tgchen_result.png', width=100)
```

```python
shuffle(draw)

display(img)
print("◼︎"*(n))
sleep(3)
clear_output(wait=True)

for i in range(n-1):
    display(img)
    print("◼︎"*(n-i))
    print(choice(cry))
    sleep(0.5)
    clear_output(wait=True)

display(img2)
print("◼︎◼︎")
sleep(3)
print('應數系50週年飛行夾克一件:', draw[0])
print('迷你超級任天堂:', draw[1])
```

# 冰雹數列

以任意的正整數 $a_1$ 開始, 產生 $a_2, a_3, \ldots$, 規則如下:

* 如果 $a_n$ 是偶數, 下一個數是 $\frac{a_n}{2}$
* 如果 $a_n$ 是奇數, 下一個數是 $3 \times a_n + 1$
    

```python
%pylab inline
def ice(a):
    if a%2 == 0:
        return a/2
    else:
        return 3*a+1
```

```python
k=23
L = [k]
while k != 1:
    k = ice(k)
    L.append(k)
plot(L)
```

```python
def iceser(k):
    L = [k]
    while k != 1:
        k = ice(k)
        L.append(k)
    plot(L)
```

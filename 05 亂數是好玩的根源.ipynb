```python
msg = input('請輸入一句話:')

if '悲傷' in msg:
    print("你的語句中有悲傷。")
elif '難過' in msg:
    print("不要想太多。")
else:
    print("嗯嗯。")
```

```python
k = np.random.randint(1, 101)
print("我想了一個 1-100 的數字, 你猜是多少。")
guess = 999

while k != guess:
    guess = int(input('你的猜測:'))
    if guess > k:
        print("太大了!")
    elif guess < k:
        print("太小了!")
        
print("答對了!!")
```

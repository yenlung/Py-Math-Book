```python
%matplotlib inline

import numpy as np
import sympy as sp
import matplotlib.pyplot as plt
```

### 3.2 Python 怎麼畫參數式?

```python
π = np.pi
```

```python
t = np.linspace(0, 2*π, 500)
```

```python
r = 3
x = r*np.cos(t)
y = r*np.sin(t)
```

```python
plt.plot(x,y);
```

### 3.3 瞭解 matplotlib 畫圖原理，畫個圓

```python
ax = plt.gca()
ax.set_aspect('equal')
plt.plot(x,y);
```

### 3.4 笛卡兒的情書

```python
t = np.linspace(0, 2*π, 500)
```

```python
r = 1 - np.sin(t)

x = r * np.cos(t)
y = r * np.sin(t)
```

```python
plt.plot(t, np.sin(t))
plt.plot(t, r);
```

```python
ax = plt.gca()
ax.set_aspect('equal')
plt.plot(x,y,c='r',lw=3)
```

```python
ax = plt.gca()
ax.set_aspect('equal')
plt.fill(x,y,c='r')
```

### 3.5 有意識的創作!

```python
t = np.linspace(0, 2*π, 500)
```

```python
r = 1 + np.sin(5*t)
```

```python
x = r * np.cos(t)
y = r * np.sin(t)
```

```python
ax = plt.gca()
ax.set_aspect('equal')
plt.plot(x,y);
```

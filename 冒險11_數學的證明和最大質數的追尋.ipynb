### 2. 判斷一個數是不是完美數

一個數的 proper divisors (真因子) 加起來剛好是自己, 就叫完美數。

```python
n = 6
egg = sp.ntheory.divisors(n)
```

```python
s = 0
for i in egg[:-1]:
  s = s + i

print(s)
```

其實 Python 知道大家很愛把一串數字加起來, 所以有標準指令...

```python
sum(egg[:-1])
```

檢查看看真因子的和 $s$ 是否就是原來的數字 $n$。

```python
n == s
```

```python
if (n==s):
  print(f"{n} 是完美數!!")
else:
  print(f"{n} 不是完美數")
```

```python
def isPerfect(n):
  egg = sp.ntheory.divisors(n)[:-1]
  s = sum(egg)
  return n==s
```

```python
list(range(10))
```

```python
list(range(6, 21))
```

```python
for i in range(5):
  print(i)
```

列出 1 到 9487 有多少個完美數。

```python
for n in range(1, 9487):
  if isPerfect(n):
    print(n)
```

```python
import networkx as nx
from networkx.drawing.nx_agraph import graphviz_layout
```

```python
G = nx.DiGraph()
```

```python
G.add_nodes_from(range(1,10))
```

```python
G.nodes
```

```python
G.add_edges_from([(4,1), (4,7), (1,6), (7,2), (2,5), (2,3), (2,8), (5,9)])
```

```python
nx.draw(G)
```

```python
G.nodes()
```

```python
nx.draw(G, with_labels=True, node_color="#A8DADC", node_size=1000, arrows=False)
```

```python
nx.is_tree(G)
```

```python
#pos=nx.graphviz_layout(G, prog='dot')
```

```python
nx.draw(G, pos=graphviz_layout(G), with_labels=True, 
        node_color="#A8DADC", node_size=1000, arrows=False)
```

```python
if G.predecessors(9):
    print(True)
```

```python
Gs = nx.DiGraph()
```

```python
Gs.add_nodes_from(['1', '2', '3', '4', '5', '6', '7', '8', '9'])
```

```python
edges = [('4', '1'),
 ('4', '7'),
 ('1', '6'),
 ('7', '2'),
 ('2', '5'),
 ('2', '3'),
 ('2', '8'),
 ('5', '9')]
```

```python
Gs.add_edges_from(edges)
```

```python
nx.draw(Gs, pos=graphviz_layout(Gs), with_labels=True, 
        node_color="#A8DADC", node_size=1000, arrows=False)
```

```python
nx.is_tree(Gs)
```
